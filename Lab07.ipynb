### Introduction

Refraction refers to waves (or in our case light waves) passing through one medium to another such as acrylic or water. Refraction includes a change in direction of the waves when passing through the media. Light refraction follows Snell’s Law. Snell’s law explains that the sin(i) = n sin r. Where i is the incidence angle, r is the refracted angle, and n is the refractive index of the media. The refractive index n is = (sini)/(sinr). In this experiment, we will be measuring the angles of light refraction through two media, plastic and water. We will be recording our data and proving Snell’s law with support of the Snell’s law Formula.

### Experimental Plan

For our Experiment, we will be measuring the angles of light refraction through water and acrylic. We will set this up by using a polar plot (circular paper angle guide).

We will start by placing each material in the center of the polar plot. We will then precisely place the laser at the correct angle and in such a way that it directly hits the material in the center. We will then use the guide from the polar plot to record the angle of the reflected beam of light. See the figure below for reference.

We will place the laser at angles from 10° to 80° with increments of 10°. We will then measure the angles of the refracted light beams.

Using the data collected, we will calculate the refractive indexes of the different materials.

### Data and Observations

#### Acrylic

##### Python code using Pandas to produce data table.

```python
import pandas as pd
from numpy import sin, cos, linspace, array, sqrt
from math import pi

# Measured and calculated values for the acrylic sample

δθ = 0.5 * (pi/180)                           # Enter the uncertainty in your angle measurement in radians
i_a = array([10,20,30,40,50,60,70,80],float)    # Enter the incident angles separated by commas in degrees
r_a = array([7.5,13.5,19.5,25.5,30.5,35.5,39,41.5],float)    # Enter the refracted angles separated by commas in degrees
r_rad_a = (pi/180) * r_a                          # Convert to radians
i_rad_a = (pi/180) * i_a                          # Convert to radians

sin_r_a = sin(r_rad_a)
sin_i_a = sin(i_rad_a)

n_a = sin_i_a/sin_r_a                               # index of refraction. Use equation 3.30

cot = lambda θ:cos(θ)/sin(θ)

frac_un_r_a = cot(r_rad_a)*δθ               # fractional uncertainty of sin r. Use equation 3.32
frac_un_i_a = cot(i_rad_a)*δθ               # fractional uncertainty of sin i. Use equation 3.32

frac_n_a = sqrt((frac_un_i_a/sin_i_a)**2 + (frac_un_r_a/sin_r_a)**2) # fractional uncertainty of n. Use equation 3.31


un_sinr_a = frac_un_r_a * sin_r_a
un_sini_a = frac_un_i_a * sin_i_a

print('Table 1.  Measured and caluclated values for the acrylic sample.')
data_a = {'r (deg)': r_a, 'i (deg)': i_a, 'sin r': sin_r_a, 'sin i': sin_i_a, 'n': n_a, \
        'δsin r/|sin r|': frac_un_r_a, 'δsin i/|sin i|': frac_un_i_a,'δn/n': frac_n_a,'δsin r':un_sinr_a,'δsin i':un_sini_a}
acrylic_data = pd.DataFrame(data_a)
acrylic_data.round(3)
```
